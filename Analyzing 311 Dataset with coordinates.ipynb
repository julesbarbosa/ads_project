```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_311 = pd.read_csv('merged/Type/data_2012_2013.csv')
```

```python
df_311.head()
```

```python
df_311['Incident Zip'].unique()
```

```python
def zipcode_cleaner(x):
    x = str(x)
    x = x[:5]
    if '.' in x:
        x = '0' + x
    elif len(x)<5:
        return np.nan
    return x[:5]
```

```python
df_311['Incident Zip'] = df_311['Incident Zip'].apply(lambda x: zipcode_cleaner(x))
```

```python
zip_dict = dict.fromkeys(df_311['Incident Zip'].unique().tolist())
```

```python
for z in df_311['Incident Zip'].unique().tolist():
    try:
        zip_dict[z]=[search.by_zipcode(z).lng, search.by_zipcode(z).lat]
    except:
        print(z)
```

```python
zip_dict
```

```python
zipcode_coord = pd.DataFrame(zip_dict).transpose().rename(columns={0:'lng',1:'lat'})
```

```python
zipcode_coord.shape
```

```python
zipcode_coord = zipcode_coord.loc[zipcode_coord['lng'].isnull()==False]
```
